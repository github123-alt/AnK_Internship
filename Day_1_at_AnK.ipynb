```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population = pd.read_csv('/population.csv')
```

```python
population.head()
```

```python
population.info()
```

```python
countries = ['Nepal','India','China']

filtered_population = population[
    (population['Country Name'].isin(countries)) &
    (population['Year'].between(2002, 2020))
    ]

print(filtered_population)
```

```python
display(population['Value'].mean())
display(population['Value'].min())
display(population['Value'].max())
```

```python
population['GrowthRate'] = population.groupby('Country Name')['Value'].pct_change() * 100
average_growth = population['GrowthRate'].mean()
print("Average growth rate across all countries:", average_growth)
```

```python

plt.figure(figsize=(10,6))
for country, group in population.groupby('Country Name'):
    plt.plot(group['Year'], group['Value'], label=country)

plt.title('Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
year = 2010
population_year = population[population['Year'] == year]

plt.figure(figsize=(8,5))
plt.bar(population_year['Country Name'], population_year['Value'])
plt.title(f'Population by Country in {year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
country = 'Nepal'
population_country = population[population['Country Name'] == country]

plt.figure(figsize=(8,5))
plt.hist(population_country['Value'], edgecolor='red')
plt.title(f'Population Distribution for {country}')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
